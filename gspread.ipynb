```python
import gspread
from oauth2client.service_account import ServiceAccountCredentials
from gspread_dataframe import set_with_dataframe
import pandas as pd
```

```python
#jsonファイルを使って認証情報を取得
scope = ['https://spreadsheets.google.com/feeds','https://www.googleapis.com/auth/drive']
c = ServiceAccountCredentials.from_json_keyfile_name('/Users/kii/lang/python/gspread/local-now-340614-96ac1e456268.json', scope)
```

```python
#認証情報を使ってスプレッドシートの操作権を取得
gs = gspread.authorize(c)
```

```python
#共有したスプレッドシートのキー（d/<ここがキー>/edit）を使ってシートの情報を取得
SPREADSHEET_KEY = '1n_NaF1GOaB6vgquY1eyAnnlamle4SIYoGwE1mASoFj8'

workbook = gs.open_by_key(SPREADSHEET_KEY)
#取得したいシート名
worksheet = workbook.worksheet("新規ページ")
```

```python
#シートをデータフレームに格納
df = pd.DataFrame(worksheet.get_all_values())
df
```

```python
#[0]にある値をカラム名に修正
df.columns = df.iloc[0]
df = df.drop(df.index[[0]])
df
```

```python
#データフレームの内容をワークシートにセット、反映
workbook.add_worksheet(title="新規ページ", rows=50, cols=10)
```

```python
#ページにデータフレームを書き込み
set_with_dataframe(workbook.worksheet("新規ページ"), df, include_index=True)
```
